##### Q1. Design and analyze algorithms for following, based on their time complexity : Linear Search and Binary Search

##### Import Library

```python
import matplotlib.pyplot as plt
import math
```

##### Sample Array Generation

```python
rangeLim = 300
x = [i+1 for i in range(rangeLim)]
print(x)
```

##### Analysis of Algorithm(Binary Search)

```python
it = []
for i in range(rangeLim):
    c = 0
    n = x[i]
    l = 0
    r = len(x)-1
    while(l<=r):
        mid = (l+r)//2
        if(x[mid]==n):
            it.append(c)
            break
        elif(x[mid]<n):
            l = mid+1
        else:
            r = mid-1
        c+=1
print(it)
```

```python
logx = [math.log2(x[i]) for i in range(rangeLim)]
plt.scatter(it,x)
plt.plot(logx,x, color='red')
plt.ylim([0,rangeLim])
plt.xlim([0,max(it)+1])
plt.title("Binary Search")
plt.xlabel("Iterations")
plt.ylabel("Number of elements")
```

##### Analysis of Algorithm(Linear Search)

```python
it2 = []
for j in range(rangeLim):
    l = x[j]
    for i in range(len(x)):
        if(x[i]==l):
            it2.append(i)
print(it2)
```

```python
plt.plot(it2,x)
plt.ylim([0,rangeLim])
plt.xlim([0,max(it2)+1])
plt.title("Linear Search")
plt.xlabel("Iterations")
plt.ylabel("Number of elements")
```
